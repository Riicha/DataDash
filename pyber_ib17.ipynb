```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
###import seaborn as sns
###from PIL import Image
```

```python
# Take in all of our wrestling data and read it into pandas
city_data_file = "raw_data/city_data.csv"
ride_data_file = "raw_data/ride_data.csv"

city_data_df = pd.read_csv(city_data_file)
ride_data_df = pd.read_csv(ride_data_file)
```

```python
# Merge two datasets on "City" so that no data is lost
combined_city_ride_df = pd.merge(city_data_df, ride_data_df, how='inner', on='city')
combined_city_ride_df.head(30)
```


# Pyber Ride Sharing Data (2016)

```python
# Create a Scatter Plot based on Pyber Ride Data (2016)
    
# Separate Combined City Ride dataset into 3 parts/datasets by "type"
rural_city_data_df = combined_city_ride_df[combined_city_ride_df['type'] == "Rural"]
###print(rural_city_data_df['ride_id'].count())

suburban_city_data_df = combined_city_ride_df[combined_city_ride_df['type'] == "Suburban"]
###print(suburban_city_data_df['ride_id'].count())

urban_city_data_df = combined_city_ride_df[combined_city_ride_df['type'] == "Urban"]
###print(urban_city_data_df['ride_id'].count())

# Get the values for each part of the scatter plot by city "type"

# Within "Rural" dataset type, get valuess for the scatter plot using GroupBy by "city"

# Using GroupBy in order to separate combined city ride data into fields by "city" values
rural_grouped_by_city_df = rural_city_data_df.groupby(['city'])

# Get values for X axis
rural_num_rides = rural_grouped_by_city_df['ride_id'].count()
###print(rural_num_rides)

# Get values for Y axis
rural_avg_fares = rural_grouped_by_city_df['fare'].mean()
###print(rural_avg_fares)

# To determine Circle size
rural_driver_count = rural_grouped_by_city_df['driver_count'].mean()
###print(rural_driver_count)
```

```python
# Within "Suburban" dataset type, get valuess for the scatter plot using GroupBy by "city"

# Using GroupBy in order to separate combined city ride data into fields by "city" values
suburban_grouped_by_city_df = suburban_city_data_df.groupby(['city'])

# Get values for X axis
suburban_num_rides = suburban_grouped_by_city_df['ride_id'].count()
###print(suburban_num_rides)

# Get values for Y axis
suburban_avg_fares = suburban_grouped_by_city_df['fare'].mean()
###print(suburban_avg_fares)

# To determine Circle size
suburban_driver_count = suburban_grouped_by_city_df['driver_count'].mean()
###print(suburban_driver_count)
```

```python
# Within "Urban" dataset type, get valuess for the scatter plot using GroupBy by "city"

# Using GroupBy in order to separate combined city ride data into fields by "city" values
urban_grouped_by_city_df = urban_city_data_df.groupby(['city'])

# Get values for X axis
urban_num_rides = urban_grouped_by_city_df['ride_id'].count()
###print(urban_num_rides)

# Get values for Y axis
urban_avg_fares = urban_grouped_by_city_df['fare'].mean()
###print(urban_avg_fares)

# To determine Circle size
urban_driver_count = urban_grouped_by_city_df['driver_count'].mean()
###print(urban_driver_count)
```

```python
# Creating Pyber Ride Sharing Data (2016) Scatter Plot

# Clear space for the scatter plot
plt.close(1)
# Organize the layout for the scatter plot
plt.title("Pyber Ride Sharing Data (2016)", Fontsize= 14 )
# Labels for the scatter plot circles by city "type"
labels = ["Rural", "Suburban", "Urban"]
# Colors for the scatter plot circles depending on city "type"
colors = ["gold","lightskyblue", "lightcoral"]

# Define X and Y of the scatter plot
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")

# Define X and Y limits
plt.xlim(0, 45 )
plt.ylim(15, 40 )

# Writing data to the same Scatter Plot 3 times, for each city "type"( "Rural", "Suburban" "Urban")
plt.scatter( rural_num_rides, rural_avg_fares, s=20*rural_driver_count, c="gold", 
            edgecolor="black", linewidths=1, alpha =0.75, marker='o', label="Rural")
plt.scatter( suburban_num_rides, suburban_avg_fares, s=10*suburban_driver_count, c="lightskyblue", 
            edgecolor="black", linewidths=1, alpha=0.75, marker='o', label="Suburban")
plt.scatter( urban_num_rides, urban_avg_fares, s=10*urban_driver_count, c="lightcoral", 
           edgecolor="black", linewidths=1, alpha=0.75, marker='o', label="Urban")
 
# Create the legend of the Scatter plot
legend = plt.legend(fontsize="small", loc="best", title="City Type")
plt.tight_layout()

plt.grid(True)

# Write comments on the right of the grid
plt.text(30,15.5,"Note:\nCircle size correlates with\ndriver count per city.")

# Save the figure with the Scatter Plot
plt.savefig("PyberRideDataScatterPlot.pdf")

# Show the Scatter Plot
plt.show()
```


# Total Fares by City Type

```python
# Get the values for each section of the pie chart
# Using GroupBy in order to separate combined city ride data into fields according to city"type" values
type_grouped_city_fares_df = combined_city_ride_df.groupby(['type'])
sum_fares = type_grouped_city_fares_df['fare'].sum()
###print (sum_fares)
```

```python
# Creating "% of Total Fares by City Type" PIE CHART
# Clear the pie chart space
plt.close(1)

# Labels for the sections of our pie chart
#labels = ["Rural", "Suburban", "Urban"]
labels = type_grouped_city_fares_df.sum().index
###print( labels)
# The colors of each section of the pie chart
colors = ["gold","lightskyblue", "lightcoral"]

# Tells matplotlib to seperate the "Urban" section from the others
explode = (0, 0, 0.2)

# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sum_fares, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

# Tells matplotlib that we want a pie chart with equal axes
plt.title("% of Total Fares by City Type", Fontsize = 14)
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("FaresByCityType.pdf")
plt.show()
###plt.tight_layout()

### Another way to display the otput chart, if Matplotlab is buggy
### file2 = "FaresByCityType.pdf"
### display(Image.open(file2))
```
